```python
import os

# Path to the raw_data folder
raw_data_path = '../data/raw_data'

# List all items in the directory
items = os.listdir(raw_data_path)

# Filter only directories
folders = [item for item in items if os.path.isdir(os.path.join(raw_data_path, item))]

# Count the folders
num_folders = len(folders)

print(f"Number of folders in {raw_data_path}: {num_folders}")
print("Folders:", folders)
```
