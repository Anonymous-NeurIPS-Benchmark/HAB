```python
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv(override=True)

test_models = {
     'llama3_3': ('meta/llama-3.3-70b-instruct',          'NVIDIA_API_KEY_LLAMA33'),
     'mamba':    ('mistralai/mamba-codestral-7b-v0.1',               'NVIDIA_API_KEY_MAMBA'),
     'phi4':     ('microsoft/phi-3-medium-128k-instruct', 'NVIDIA_API_KEY_PHI'),
 }

for model_key, (model_id, api_key_name) in test_models.items():
    try:
        client = OpenAI(
            base_url='https://integrate.api.nvidia.com/v1',
            api_key=os.environ[api_key_name]
        )
        resp = client.chat.completions.create(
            model=model_id,
            messages=[{'role': 'user', 'content': 'Say yes in one word.'}],
            max_tokens=10,
            stream=False
        )
        print(f'✅ {model_key} ({model_id}): {resp.choices[0].message.content.strip()}')
    except Exception as e:
        print(f'❌ {model_key}: {str(e)[:100]}')
```
